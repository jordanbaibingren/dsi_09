# Project 2 - Ames Housing Data and Kaggle Challenge

The aim of the project is to create a regression model based on the Ames Housing Dataset, that will predict the price of a house at sale.

```python
# Import relevant libraries

import numpy as np
import pandas as pd
import patsy

from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
# Load training and testing data
df_train = pd.read_csv('./datasets/train.csv')
df_test = pd.read_csv('./datasets/test.csv')
df = pd.concat([df_train, df_test], sort=False)


df.columns = df.columns.str.replace(' ', '_')
df.columns = map(str.lower, df.columns)
```

```python
# Identify the variables with missing values
count_nan = len(df) - df.count()
count_nan[count_nan>0]
```

```python
df.shape
```

**Imputation with Zeros:**
Quantifiable numerical features with missing values should be imputed with 0s as the houses likely did not contain the installation.

```python
# Fill empty cells with zeros for selected features
def fill_zero(features, data):
    for s in features:
        data[s] = data[s].fillna(0)
    return data

df = fill_zero(['lot_frontage', 'mas_vnr_area', 'bsmtfin_sf_1', 'bsmtfin_sf_2', 
                'bsmt_unf_sf', 'total_bsmt_sf', 'bsmt_full_bath', 'bsmt_half_bath', 'garage_yr_blt',
                'garage_cars', 'garage_area','electrical'], df)
```

**Imputation with NA:** Qualitative features with missing values should be imputed with NA as the houses likely did not contain the installation.

```python
# Fill empty cells with NA for selected features
def fill_NA(features, data):
    for s in features:
        data[s] = data[s].fillna('NA')
    return data

df = fill_NA(['alley', 'bsmt_qual', 'bsmt_cond', 'bsmt_exposure', 'bsmtfin_type_1',
               'bsmtfin_type_2', 'fireplace_qu', 'garage_type', 'garage_finish', 'garage_qual', 'garage_cond', 'pool_qc', 'fence', 'misc_feature'], df)
```

**Imputation with None:** Qualitative feature 'mas_vnr_type' missing values should be imputed with None as the houses likely did not contain the installation.

```python
# Fill empty cells with none for selected features
df['mas_vnr_type'] = df['mas_vnr_type'].fillna('None')
```

**Identifying and removing outliers:** Outliers can be identified by plotting the variable against our target variable.

```python
ax = df.plot.scatter(x='gr_liv_area', y='saleprice')
```

```python
# Removing outliers
df = df.drop(df[(df['gr_liv_area']>4000) & (df['saleprice'].notna())].index)
df.shape
```

**Combining features:** Many features in the list were measuring similar qualitities. These could either be combined (e.g. number of bathrooms), dropped (e.g. surface area of each floor vs total surface area) or integrated (e.g. compute age of house using year of sale and year of remodelling).

```python
# Sum all bath numbers
df['total_bath'] = df[['full_bath','half_bath','bsmt_full_bath','bsmt_half_bath']].sum(axis = 1, skipna = True)

# Create variable for age at sale
df['age'] = df['yr_sold'] - df['year_remod/add']
df['garage_age'] = df['yr_sold'] - df['garage_yr_blt']
```

## Pre-processing
- One-hot encode categorical variables.
- Train/test split your data.
- Scale your data.
- Consider using automated feature selection.

**Converting features:** Ordinal features were converted from object type variables to int. Categorical variables were One-hot Encoded.

```python
# Convert ordinal variables
df['utilities_o'] = df['utilities'].replace({'NA':0,'ELO':1,'NoSeWa':2,'NoSewr':3,'AllPub':4})
df['land_slope_o'] = df['land_slope'].replace({'NA':0,'Sev':1,'Mod':2,'Gtl':3})
df['heating_qc_o'] = df['heating_qc'].replace({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
df['electrical_o'] = df['electrical'].replace({'NA':0,'Mix':1,'FuseP':2,'FuseF':3,'FuseA':4,'SBrkr':5})
df['kitchen_qual_o'] = df['kitchen_qual'].replace({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
df['functional_o'] = df['functional'].replace({'NA':0,'Sal':1,'Sev':2,'Maj2':3,'Maj1':4,'Mod':5,'Min2':6,'Min1':7,'Typ':8})
df['fireplace_qu_o'] = df['fireplace_qu'].replace({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
df['garage_qual_o'] = df['garage_qual'].replace({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
df['garage_finish_o'] = df['garage_finish'].replace({'NA':0,'Unf':1,'RFn':2,'Fin':3})
df['garage_cond_o'] = df['garage_cond'].replace({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
df['pool_qc_o'] = df['pool_qc'].replace({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
df['fence_o'] = df['fence'].replace({'NA':0,'MnWw':1,'GdWo':2,'MnPrv':3,'GdPrv':4})
```

```python
# Extract features of interest
df_extract = df.drop(columns=['id','pid','ms_subclass','ms_zoning','neighborhood',
                     'year_built','1st_flr_sf','2nd_flr_sf','bedroom_abvgr',
                     'kitchen_abvgr','garage_cars','misc_feature','mo_sold','condition_2',
                    'full_bath','half_bath','bsmt_full_bath','bsmt_half_bath',
                      'yr_sold','year_remod/add','garage_yr_blt','utilities','land_slope',
                      'heating_qc','electrical','kitchen_qual','functional','fireplace_qu',
                       'garage_qual','garage_finish','garage_cond','pool_qc','fence',
                        'exterior_2nd','bsmtfin_type_2'])
```

```python
fig = plt.subplots(figsize=(16,8))
ax = sns.heatmap(df_extract.corr(), annot=False)
ax.set_title('Correlation Matrix of Variables', fontsize=21)
```

```python
# Convert categorical variables
df_extract = pd.get_dummies(df_extract, dummy_na=False)
```

```python
# Splitting Test set where target variable is not defined
df_select = df_extract[df_extract['saleprice'].notna()]
df_select_test = df_extract[df_extract['saleprice'].isnull()]
```

```python
df_select_test.shape
```

**Train/Test Split:** Further split the training data into a test set and a training set with a 70-30 split. The latter is for cross-validation.

```python
# Define target variable and perform Train/test split
nc = [x for x in df_select.columns if x != 'saleprice']
X = df_select[nc]
y = np.log(df_select['saleprice'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale data
scaler = StandardScaler()
scaler = scaler.fit(X)
X_train_scale = scaler.transform(X_train)
X_test_scale = scaler.transform(X_test)
```

## Modeling
- **Establish your baseline score.**
- Fit linear regression. Look at your coefficients. Are any of them wildly overblown?
- Fit lasso/ridge/elastic net with default parameters.
- Go back and remove features that might be causing issues in your models.
- Tune hyperparameters.
- **Identify a production model.** (This does not have to be your best performing Kaggle model, but rather the model that best answers your problem statement.)
- Refine and interpret your production model.

```python
# Establish baseline score
baseline_mse = np.mean(y_test - np.mean(y_train))
print(baseline_mse)
```

```python
# Linear model
lr = LinearRegression() 
lr.fit(X_train_scale, y_train)
lr_cv_mean_r2 = cross_val_score(lr, X_train_scale, y_train, cv=5).mean()
lr_cv_mean_mse = -cross_val_score(lr, X_train_scale, y_train, cv=5, scoring='neg_mean_squared_error').mean()
```

```python
# Ridge model
r_alphas = np.logspace(0, 5, 200)
ridge_model = RidgeCV(alphas=r_alphas, store_cv_values=True)
ridge_model = ridge_model.fit(X_train_scale, y_train)

# Here is the optimal value of alpha
ridge_optimal_alpha = ridge_model.alpha_
print(ridge_optimal_alpha)

# Let's average the CV scores for each value of alpha
ridge_cv_means = [np.mean(cv_alpha) for cv_alpha in ridge_model.cv_values_.T]
```

```python
def plot_cv(alphas, cv_means, optimal_alpha, log=False):
    # alphas = list of alphas
    # cv_means = list of CV mean MSE
    # optimal_alpha
    # lr_mse
    fig = plt.figure(figsize=(12,8))
    ax = plt.gca()

    if log:
        ax.semilogx(alphas, cv_means, lw=2)
    else:
        ax.plot(alphas, cv_means, lw=2)
    ax.axvline(optimal_alpha)
    #ax.axhline(lr_mse)
    ax.set_xlabel('alpha')
    ax.set_ylabel('Mean Squared Error')
    
plot_cv(ridge_model.alphas, ridge_cv_means, ridge_optimal_alpha, log=True)
```

```python
# Lasso model
l_alphas = np.arange(0.0001, 0.01, 0.0025)
#l_alphas = np.logspace(-2, 3, 50)
lasso_model = LassoCV(alphas=l_alphas, cv=5)
lasso_model = lasso_model.fit(X_train_scale, y_train)

# Here is the optimal value of alpha
lasso_optimal_alpha = lasso_model.alpha_
print(lasso_optimal_alpha)

# Let's average the CV scores for each value of alpha
lasso_cv_means = [np.mean(cv_alpha) for cv_alpha in lasso_model.mse_path_]

plot_cv(lasso_model.alphas_, lasso_cv_means, lasso_optimal_alpha)
```

```python
# Evaluate training models
print('Baseline:',baseline_mse)
print('Linear Regression:',lr_cv_mean_mse)
print('Ridge:',min(ridge_cv_means))
print('Lasso:',min(lasso_cv_means))
```

```python
# Evaluate testing models
print('Linear regression:',lr_cv_mean_r2)
print('Linear regression test:',lr.score(X_test_scale, y_test))

ridge_opt = Ridge(alpha=ridge_optimal_alpha)
print('Ridge CV with optimal alpha:',cross_val_score(ridge_opt, X_train_scale, y_train, cv=5).mean())
ridge_opt.fit(X_train_scale, y_train)
print('Ridge test with optimal alpha',ridge_opt.score(X_test_scale, y_test, sample_weight=None))

lasso_opt = Lasso(alpha=lasso_optimal_alpha)
print('Lasso CV with optimal alpha:',cross_val_score(lasso_opt, X_train_scale, y_train, cv=5).mean())
lasso_opt.fit(X_train_scale, y_train)
print('Lasso test with optimal alpha',lasso_opt.score(X_test_scale, y_test))
```

**Identify a production model.** Lasso was chosen from the cross-validation as it had the lowest MSE out of all the models. Implementing the models using optimal hyperparameter alpha, Lasso also performed better than Ridge and Linear Regression on both training and test scores.

```python
# Production model chosen to be Lasso
X_real_test_scale = scaler.transform(df_select_test.drop(columns=['saleprice']))
X_scale = scaler.transform(X)

lasso_real_test = Lasso(alpha=lasso_optimal_alpha)
lasso_real_test.fit(X_scale, y)

y_real_test_predict = lasso_real_test.predict(X_real_test_scale)
```

```python
y_final = pd.DataFrame(np.exp(y_real_test_predict))
final_kaggle = y_final.join(df_test["Id"], how='outer')
final_kaggle = final_kaggle[['Id', 0]]
final_kaggle.columns = ['id','saleprice']
final_kaggle.to_csv('jb_submit_kaggle.csv',index=False)
```
